## Importando as Bibliotecas Necess치rias

```python
import pandas as pd
import numpy as numpy
import matplotlib.pyplot as pyplot
import seaborn as sns
%matplotlib inline

from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
```

## Carregando o dataset

```python
pathTeach='00-Bases-de-dados/Professores- X-Niveis/pisa-professores-niveis.csv'
professoresDF =  pd.read_csv(pathTeach)
professoresDF = professoresDF.drop(['Unnamed: 0'],axis=1) 
```

```python
from  sklearn.model_selection  import train_test_split
```

## Separando as vari치veis x e y

```python
x = professoresDF.drop(['CNT','nivelMatematica','nivelCiencias'],axis=1) # 
y = professoresDF['nivelCiencias'] #  seleciona somente o que se quer prever 

```

## Feature Selection

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import RFE
dt= RandomForestClassifier()
rfe = RFE(dt)
fit = rfe.fit(x,y)
cols = fit.get_support(indices=True)
x=professoresDF.iloc[:,cols]

```
